```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("markerless data mock.xlsx")
```

```python
df.head()
```

```python
sns.boxplot(data=df, x="Blind zones", y="Make Model")
```

```python
conditions = [
    (df['Model Year']<2000), 
     (df['Model Year']<2005),
     (df['Model Year']<2010),
     (df['Model Year']<2015),
     (df['Model Year']<2020),
     (df['Model Year']<2025)
    ]

# create a list of the values we want to assign for each condition
values = ['1995', '2000', '2005', '2010', '2015', '2020']

# create a new column and use np.select to assign values to it using our lists as arguments
df['generation'] = np.select(conditions, values)
df = df.dropna()
df = df.sort_values(['generation'], ascending = True)
# display updated DataFrame
df.head()
```

```python
df_honda_crv = df[df['Make Model'].isin(["Honda CR-V"])]

conditions_hondacrv = [
     (df_honda_crv['Model Year']<2002),
     (df_honda_crv['Model Year']<2007),
     (df_honda_crv['Model Year']<2012),
     (df_honda_crv['Model Year']<2017),
     (df_honda_crv['Model Year']<2025)
    ]

# create a list of the values we want to assign for each condition
values_hondacrv = ['Generation 1', 
          'Generation 2', 
          'Generation 3', 
          'Generation 4', 
          'Generation 5']

# create a new column and use np.select to assign values to it using our lists as arguments
df_honda_crv['honda crv generation'] = np.select(conditions_hondacrv, values_hondacrv)
# display updated DataFrame
df_honda_crv.head()
```

```python
ax = sns.boxplot(data=df_honda_crv, y="Blind zones", x="honda crv generation")

ax.set(xlabel='Vehicle Generations',
       ylabel='Total Blind Zone Area Around Vehicle (M^2)',
       title='Honda CR-V Blind Zone Area 1995-2020')

#display barplot
plt.show()
```

```python
ax = sns.boxplot(data=df, y="Blind zones", x="generation", hue = 'Body Type')

ax.set(xlabel='Vehicle Generations',
       ylabel='Total Blind Zone Area Around Vehicle (M^2)',
       title='Blind Zone Area by Body Type 1995-2020')

#display barplot
plt.show()
```
